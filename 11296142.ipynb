```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load the dataset into a Pandas DataFrame
data = pd.read_csv('red_wine.csv')
```

```python
# Display the first 10 rows of the dataset
data.head(10)
```

```python
#Display the last 20 rows of the dataset
data.tail(20)
```

```python
# Check basic information about the dataset
data.info()
```

```python
# Obtain statistical information on the dataset
data.describe()
```

```python
# Plot histograms for numerical columns
data.hist(figsize=(10, 10))
plt.tight_layout()
plt.show()
```

```python
# Generate a correlation matrix
corr_matrix = data.corr()

# Plotting the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

## Exploratory Data Analysis (EDA) Report

### Histograms of Wine Properties:

Histograms were generated to visualize the distributions of various wine properties. These histograms offer insights into the frequency and spread of individual properties:

- *pH Distribution:* The histogram displays a roughly normal distribution of pH values, centered around the mean value 3.3111 with a uniform spread.



### Correlation Matrix Heatmap:

A correlation matrix heatmap was created to understand relationships between different wine properties:

- *Correlation Insights:* The heatmap showcases correlations among properties. For instance, there appears to be a strong positive correlation between Citric Acid and Fixed Acidity, indicating a potential relationship worth exploring further.
